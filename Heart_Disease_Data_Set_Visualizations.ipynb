Heart Disease Diagnostic Simulation 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
file = './Desktop/processed_cleveland_data.txt'
df = pd.read_csv(file)
columns = ['Age', 'Sex', 'Chest_Pain_Type', 'RestBP', 'Chol', 'FBS', 'RestECG',
  'Thalach', 'Exang', 'OldPeak', 'Slope', 'Ca', 'Thal', 'Pred_Diagnosis']
df.columns = columns
print(df)

plt.figure(figsize =(20,20))

#This boxplot visual shows the distribution of patient ages organized by quartiles. There are no fliers,
#so there are no ages that are 1.5xIQR and require specific investigation. Contextually, 
#if there was such an age (for example a very low age), the heart disease may have more to do with a congenital
#or genetic defect than with the 14 measured attributes. The middle 50% of ages used for this program
#range from a little over 45 years to a little under 65 years.
plt.subplot(2,2,1)
plt.boxplot(df['Age'])
plt.title('Age Range')
plt.ylabel('Age')
plt.ylim([20,80])

#This histogram shows resting blood pressure comparisons for patients admitted to the hospital. 
#It illustrates that the majority of bp values were in the 120-140 mmHg range. The majority 
#of admitted patients were in the prehypertension range. 
plt.subplot(2,2,2)
plt.hist(df['RestBP'], color = 'red')
plt.ylim([0,80])
plt.title('Resting Blood Pressure Upon Hospital Admission (mm Hg)')
plt.xlabel('Resting BP')

#This comparison of two histograms illutstrates that asymptomatic patients were more numerous than any other 
#group and a predicted diagnosis of 0 was more common than any other diagnosis. The least difference
#was between patients who presented with atypical angina and patients who received a predicted
#diagnosis score of 2. This warrants further investigation.
plt.subplot(2,2,3)
plt.hist(df['Chest_Pain_Type'], color = 'blue', label = 'Chest Pain Type')
plt.hist(df['Pred_Diagnosis'], color = 'green', label = 'Predicted Diagnosis')
plt.title('Comparison of Chest Pain Type and Predicted Diagnosis')
plt.legend(loc = 'upper center')

#This scatterplot in fact illustrates medical knowledge that there is a positive correlation between
#an increase in cholesterol and an increase in blood pressure. The majority of patients presented
#with 200-300 mg/dl Cholesterol reading and a 120-140 prehypertensive resting blood pressure.
plt.subplot(2,2,4)
plt.scatter( x=df['Chol'], y=df['RestBP'], marker = 'x', s=10, color = 'red')
plt.title("Resting Blood Pressure vs. Cholesterol")
plt.ylabel('Resting Blood Pressure (mm Hg)')
plt.xlabel('Cholesterol (mg/dl)')


```
